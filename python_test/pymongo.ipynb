```python
from pymongo import MongoClient  #logical query selector

client = MongoClient()
database = client["test"]
collection = database["news"]

count = 0
#查詢滿足多重條件的語法是{"$and":[{條件1},{條件2},{條件3},...]}
#另有$or, $not, $nor
#$all在pymongo已廢止, 因為與$and功能重複

for post in collection.find(
    {"date":"20160827"}): 
    print(post["title"])
    count +=1
print("----------------")
print(count)
client.close()
```
