## 用 [or語法] 搜尋兩天內的PTT文章反而速度很慢

```python
import re 
import time
tic = time.clock()

tag=''
date = '20160923'
date_1 = '20160921'
all_title = []
# print tag
for post in collection.find({"date":date}):
    news_title = post['title'].strip()
    news_date = post['date']
    url = post['url']
    comp = post['comp']
    
    for post in collection2.find(
        {"$or":[                   
            {"date":date},
#             {"date":date_1},
        ],
        "category":"新聞",
        "title":{"$regex":news_title}}):
        print "----------------新聞標題---------------"+news_title.encode('utf-8')
        print "----------------新聞日期---------------"+news_date.encode('utf-8')
        print "----------------新聞日期---------------"+url.encode('utf-8')
        print "----------------報社---------------"+comp.encode('utf-8')
        print post['title'],post['url'],post['date']
        
toc = time.clock()
print "執行時間:",(toc - tic)
```
